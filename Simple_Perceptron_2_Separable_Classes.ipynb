```python
import numpy as np
def perceptron(x,y):
  w=np.array([0,0])
  #w = np.zeros(x.shape[1])   # works for any number of features
  l=1
  b=0
  k,q=x.shape
  for r in range(10):
    for i in range(k):
      summ=np.dot(w,x[i])+b
      if summ >=0:
        sum1=1
      else:
        sum1=-1
      if sum1!= y[i]:
        w=w+l*x[i]*y[i]
        b=b+l*y[i]
  return w,b
```

```python

x=np.array([[2,2],[4,4],[-2,-2],[0,-4]]) #training
y=np.array([1,1,-1,-1])
weight,bias=perceptron(x,y)
test = np.array([[3,3],[5,1],[-3,-1],[1,-3]]) #testing
d=test.shape[0]
for i in range(d):
  sum2=np.dot(weight,test[i])+bias
  if sum2>=0:
    print("1")
  else:
    print("-1")
```
